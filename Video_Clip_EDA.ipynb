```python
import pandas as pd

# Load the uploaded file to examine its contents
file_path = 'clips.csv'
#read file with latin1 encoding
data = pd.read_csv(file_path, encoding='latin1')
data[:5]

```

```python
import matplotlib.pyplot as plt

# Extracting the video length column
video_lengths = data['Sum of chunk_length']

# Creating a histogram
plt.figure(figsize=(10, 6))
plt.hist(video_lengths, bins=10, edgecolor='black')
plt.title('Histogram of Video Lengths')
plt.xlabel('Video Length (seconds)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Histogram of Laugh Instances
laughter_instances = data['Sum of laughter_indicator']

# Creating a histogram
plt.figure(figsize=(10, 6))
plt.hist(laughter_instances, bins=10, edgecolor='black')
plt.title('Histogram of Number of Laughter Instances')
plt.xlabel('Number of Laughter Instances')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Histogram of the Ratio
ratios = data['Ratio']

# Creating a histogram
plt.figure(figsize=(10, 6))
plt.hist(ratios, bins=10, edgecolor='black')
plt.title('Histogram of Laughs/Seconds Ratios')
plt.xlabel('AVG Seconds Between Laughs')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
#Average Video Length
avg_vid = data['Sum of chunk_length'].mean()
avg_vid
```
